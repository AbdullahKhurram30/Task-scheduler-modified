## Dynamic programming 

Description of the strategy

Dynamic programming is a technique which looks to make sure that the we compute the final output by combining solutions to smaller and smaller subproblems as possible. The best method to do this in the case of a task scheduler is to compute the best possible combination that gives us the highest utility/profit value. This is done through comparing different combinations and their respective profit values and then giving the user the best possible case. The approach is to use a table to keep track of the best possible profit values. The algorithm runs over each possibility, compares each profit value with the previous one and if it is higher than the previous, it notes it down in the table or if it is lower, notes down the previous value in that position too. This way, when it reaches the end of the table, the table[n][m] value is the highest possible value. This is the final output that we use.

#### Assumptions:

The first assumption that we make with dynamic programming is that the tasks are not fixed i.e. they do not have to happen at a certain time. This assumption is there because in order to consider all the possible combinations, we need them to be free of such constraints and although fixed cases have been implemented in the case of the greedy approach, with dynamic, the fixed cases have been removed. This allows freedom of choice for the algorithm when scheduling activities and allows for the algorithm to make sure that the best combination of activities is given to the user.

The second assumption is that the person who is inputting the activities has a good sense of the profit that they expect to get out of the activity as well as the cost of each activity. Here, we have taken the cost to be the negative balance of money that each activity brings with it e.g. playing basketball is free of cost because the court is public and at walking distance so does not have a negative impression on money in our possession but the cost of going shopping has a cost which is subtracted from the profit to get the final priority which is the cost of the bus as well as the negative way in which this shopping trip will affect the bank balance. This is how it becomes a pay off between the profit of going shopping and the following happiness vs the cost of the shopping trip and whether it outweighs the negative impacts to go shopping.

The last assumption that we make is that the dynamic checking of all the possible combinations give us the optimal output in the sense of the profit value that will be the highest possible. However, this check does not account for one variable that is multi-tasking because it does not factor in two tasks being performed at the same time which means that there are cases where the maximum possible output from the dynamic check will differ and will usually be lower than the actual value observed from the activities.



```python
#Maxheap code that both algorithms use
class MaxHeapq:
    """ 
    A class that implements properties and methods 
    that support a max priority queue data structure
    
    Attributes
    ----------
    heap : arr
    A Python list where key values in the max heap are stored
    heap_size: int
    An integer counter of the number of keys present in the max heap      
    """  

    def __init__(self):       
        self.heap       = []
        self.heap_size  = 0
        
    def left(self, i):
       
        return 2 * i + 1

    def right(self, i):

        return 2 * i + 2

    def parent(self, i):
    

        return (i - 1) // 2

    def maxk(self):     

        return self.heap[0]     
    
  
    def heappush(self, key):  
        self.heap.append(Task(-1, '', 10, [], -float("inf"), 0))
        self.increase_key(self.heap_size,key)
        self.heap_size += 1 
        
    def increase_key(self, i, key): 
        if key.priority < self.heap[i].priority:
            raise ValueError('new key is smaller than the current key')
        self.heap[i] = key
        while i > 0 and self.heap[self.parent(i)].priority < self.heap[i].priority:
            j = self.parent(i)
            holder = self.heap[j]
            self.heap[j] = self.heap[i]
            self.heap[i] = holder
            i = j    
            
       
    def heapify(self, i):
        l = self.left(i)
        r = self.right(i)
        heap = self.heap
        if l <= (self.heap_size - 1) and heap[l].priority > heap[i].priority:
            largest = l
        else:
            largest = i
        if r <= (self.heap_size-1) and heap[r].priority > heap[largest].priority:
            largest = r
        if largest != i:
            heap[i], heap[largest] = heap[largest], heap[i]
            self.heapify(largest)

    def heappop(self):
        if self.heap_size < 1:
            raise ValueError('Heap underflow: There are no keys in the priority queue ')
        maxk = self.heap[0]
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self.heap_size -= 1
        self.heapify(0)
        return maxk
```

```python
#code for dynamic scheduler

class Task:
    """
    - id: Task Id   
    - description: Short description of the task   
    - duration: Duration in minutes   
    - dependencies: An array of id's of tasks that the task depends on 
    - status: Current status of the task
    - multi_tasking: Tells whether the task can be multitasked or not
    - fixed: Bool for if the task has a fixed start time or can start at any time within its urgency
    - priority: The urgency of the task 
   
    """

    def __init__(self,task_id,description,duration,dependencies, profit, cost = 0, multi_tasking=False, status="N"):
        self.id = task_id
        self.description = description
        self.duration = duration
        self.dependencies = dependencies
        self.status = status
        self.cost = cost
        self.priority = profit - cost
        self.multi_tasking = multi_tasking
    
    #The format of the display
    def __repr__(self):
        return f"{self.description} - id: {self.id}\n \tDuration:{self.duration}\n\tDepends on: {self.dependencies}\n\tMust be completed at most {self.priority} minutes after day start\n\tFixed task start: {self.fixed}\n\tStatus: {self.status}"
    
    #Definining the comparison variable which in this case is the priority
    def __lt__(self, other):
        return self.priority < other.priority

class MultiTaskScheduler_Dynamic:
    """
    A Daily Task Scheduler Using Priority Queues based on priority that allows for multi-tasking
    Attributes
    ----------
    tasks : arr
        A Python list of all the tasks we schedule
    priority_queue: arr
        The priority queue of tasks ready to execute
    fixed_tasks: arr
        The priority queue of tasks that start at a fixed time
    """ 
    
    NOT_STARTED ='N'
    IN_PRIORITY_QUEUE = 'I'
    COMPLETED = 'C'
    
    def __init__(self, tasks):
        self.tasks = tasks
        self.priority_queue = MaxHeapq()
                
    def print_self(self):
        print('Input List of Tasks')
        for t in self.tasks:
            print(t)            
            
    def remove_dependency(self, task_id):
        """
        Input: list of tasks and id of the task completed
        Output: lists of tasks with completed task removed
        """
        for t in self.tasks:
            if t.id != task_id and task_id in t.dependencies:
                t.dependencies.remove(task_id)           
            
    def get_tasks_ready(self):
        """ 
        Input: list of tasks
        Output: list of tasks that are executable right now(no dependencies)
        """
        for task in self.tasks:
            if task.status == self.NOT_STARTED and not task.dependencies: 
                task.status = self.IN_PRIORITY_QUEUE 
                self.priority_queue.heappush(task)

    def check_unscheduled_tasks(self):
        """
        Input: list of tasks 
        Output: boolean (checks the status of all tasks and returns True if at least one task has status = 'N'
        """
        for task in self.tasks:
            if task.status == self.NOT_STARTED:
                return True
        return False   
    
    def format_time(self, time):
        return f"{time//60}h{time%60:02d}"
    

    
    def get_combinations_ready(self):
        times = []
        profits = []
        for task in self.tasks:
            times.append(task.duration)
            profits.append(task.priority)
        
        return times, profits
    
    def best_possible_combination(self, time = 960):
        times = self.get_combinations_ready()[0]
        profits = self.get_combinations_ready()[1]
        n = len(profits)
        K = [[0 for x in range(time + 1)] for x in range(n + 1)]
        
        for i in range(n + 1):
            for w in range(time + 1):
                if i == 0 or w == 0:
                    K[i][w] = 0
                elif times[i - 1] <= w:
                    K[i][w] = max(profits[i - 1] + K[i - 1][w - times[i - 1]], K[i - 1][w])
                else:
                    K[i][w] = K[i - 1][w]
                    
        return K[n][w]

    
    #method that checks for and executes multi-tasking
    def multi_tasker(self, task, current_time, profit):
        if (task.multi_tasking):
            temporary = []
            while(self.priority_queue.heap_size > 0 and (not self.priority_queue.maxk().multi_tasking or (self.priority_queue.maxk().duration))):
                temporary.append(self.priority_queue.heappop())

            # if there are tasks in the priority queue after the while loop is done, we can multi-task using them
            if self.priority_queue.heap:
                multitask = self.priority_queue.heappop()
                if ((current_time + task.duration)//60 < 24) and ((current_time + multitask.duration)//60 < 24):
                    print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing tasks {task.id} and {multitask.id} that take {task.duration} and {multitask.duration} mins respectively")
    
                #Increment the current time with the duration of the larger task(duration wise) as that task will still
                #be in process when the first task is completed.
                if task.duration < multitask.duration:
                    current_time += task.duration
                    if ((current_time + task.duration)//60 < 24) and ((current_time + multitask.duration)//60 < 24):
                        profit += task.priority + multi_task.priority 
                        print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")
                        current_time += multitask.duration - task.duration
                        print(f"‚úÖ Completed Task {multitask.id} - '{multitask.description}' at time {self.format_time(current_time)}\n")
        
                else:
                    current_time += multitask.duration
                    if ((current_time + task.duration)//60 < 24) and ((current_time + multitask.duration)//60 < 24):
                        profit += task.priority + multi_task.priority
                        print(f"‚úÖ Completed Task {multitask.id} - '{multitask.description}' at time {self.format_time(current_time)}\n")
                        current_time += task.duration - multitask.duration
                        print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")
                        
                self.remove_dependency(task.id)
                self.remove_dependency(multitask.id)
                task.status = self.COMPLETED
                multitask.status = self.COMPLETED

            #if the priority queue is empty after the while loop:
            #none of the tasks can be multitasked 
            #within the time between the current time and the start of the next fixed task
            #so we just perform the initial task without multi-tasking
            else:
                if (current_time + task.duration)//60 < 24:
                    print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration
                    print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")  
                    profit += task.priority
                self.remove_dependency(task.id)
                task.status = self.COMPLETED

            #once we've carried out the task:
            #we move all the other tasks to the priority queue from the temporary list we had added them to
            for i in temporary:
                self.priority_queue.heappush(i)
            
        #if the task isn't multi-taskable we treat it as a regular task without running it through further comparisons
        else: 
            if (current_time + task.duration)//60 < 24:
                print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                current_time += task.duration
                print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")
                profit += task.priority
            self.remove_dependency(task.id)
            task.status = self.COMPLETED
           
        #return the current time
        return current_time, profit
    
    
    
    def run_task_scheduler(self, starting_time = 480):
        print(f"Maximum possible profit = {self.best_possible_combination()}")
        current_time = starting_time
        profit = 0
        while self.check_unscheduled_tasks() or self.priority_queue.heap:
            self.get_tasks_ready()
            if current_time//60 > 24:
                print("End of the day. No more tasks possible today")
                break
            if self.priority_queue.heap_size > 0:  
                if self.priority_queue.maxk().duration:  
                    temporary = []
                    while(self.priority_queue.heap_size > 0):
                        temporary.append(self.priority_queue.heappop())   
                    task = temporary.pop(0)  
                        
                    # calling the multi_task method and update the current time to the 
                    #new time once the multi-task method is done.
                    current_time, profit = self.multi_tasker(task, current_time, profit)
        
                    for i in temporary:
                        self.priority_queue.heappush(i)
                        
                #if the current task can be done before the next fixed task but can not be multi-tasked with another
                #task, we just perform it like a normal task.
                else:
                    task = self.priority_queue.heappop()    
                    current_time, profit = self.multi_tasker(task, current_time, profit)
                  
        total_time = current_time - starting_time             
        print(f"üèÅ Completed all planned tasks in {total_time//60}h{total_time%60:02d}min")
        print(f"Profit achieved through this combination of activities = {profit}")
```

#### Time Complexity

The problem with these task schedulers as we can see is that the input that we give them is very hard to generate by hand even and doing it by random generation and trying to come up with a way to make an asymptotic notation would have been in vain. Thus, in order to discuss the time complexity of the dynamic multi tasker, we rely on time recurrence relations with respect to the logical reasoning of how the algorithm works.

Time recurrence calculation

Costs of various parts of the algorithm

Best_possible_combination() = O(n^2) as O(n) for the rows and O(n) for the columns which make it equal to O(n^2)

Multi_tasker() = O(n) x O(n - 1) as the function is run on all the tasks and then run again on the tasks minus the task that was performed. This simplifies down to O(n^2) - O(n).

Run_taskscheduler() = O(n) x O(n - 1) as it runs the task and then runs the method again on the tasks that are left which also simplifies down to O(n^2) - O(n).

The methods within the class that are called such as get_tasks_ready(), get_combinations_ready(), and get_unscheduled_tasks() take O(n) time which repeated for a constant number of times will lead to C*O(n)

The runtime of the heappush() method is O(n log n) as it iterates the number of elements in the list it has to push, n and the time of insertion is log n.

The extraction of the root nodes for the sake of tasks take O(1) time and multiplied by a constant means the total time will be K*O(1)

Thus, the time recurrence equation  becomes:

T(n) = O(n^2) + O(n^2) - O(n) + O(n^2) - O(n) + C*O(n) + O(n log n) + K*O(1)

The 2*O(n) + C*O(n) will become C*O(n) as C is just a constant 

T(n) = 3 O(n^2) + O(n log n) + C*O(n) + K*O(1)

As we reach asymptotic notation in theory, only the O(n^2) will be important and the others can be disregarded as their growth rate is too small for significant impact. Moreover, the constant, 3 which is the multiple of O(n^2) can also be disregarded as it is just a constant multiple.

Thus, the final time recurrence equation becomes:

T(n) = O(n^2)

This means that for each additional task added to the task scheduler, the time required to run the algorithm will grow exponentially over time. This also makes sense in the sense of the intuitive reasoning as each new task will bring in a whole row and column to the table we are using which makes the growth "n^2"



### Space Complexity

In terms of space complexity, the dynamic algorithmic approach requires more space as compared to the greedy one because it makes use of the table and thus has to store all the values in the memory as it goes along calculating all of them. This means that if we are to compare the space complexity of the algorithms, dynamic would require more space.

In terms of bytes, as each of this space in the table is an additonal integer that is stored in the table, the additional space required for the table in the dynamic approach would be 4*n^2 as each integer takes up 4 bytes.

## Greedy algorithm 

In the greedy algorithmic approach, the main property that we look to satisfy is the greedy property itself which means that if we make the best choices available to us at each point, we will make an optimal subset for the overall problem. Although this approach does not guarentee us the best results for each run, it is possible to get an optimal result in certain cases.

I have taken two approaches to this greedy algorithm.

The overarching approach of the algorithm is that it takes into account the tasks available with the highest priority, checks if they are a viable choice to be implemented i.e. they do not cross the upper bound for time that we have set and then, the task is done if it passes the checks. Thus, at each choice, we make the best available choice that is viable as if the first task is not viable, we will move on to the next one with the highest priority.

The two different approaches vary in terms of how they use the priority values. The first approach takes the priority value as a whole and compares it to the priority of others and the only use of duration is to make sure that the task is viable. The second implementation however, first divides the priority by the duration of the task to get a priority value that is per minute. Then, it compares this to the other priority per minute values to get the best possible outputs. The duration is still used when we look to see if the task is viable or not.

#### Data Strcutures uses:

The data structures that we have used in this implementation of the task scheduler are heaps, classes, and tables.

Heaps are implemented to sort the tasks based on the priority that they have and this helps to easily sort them out as heaps are very effective and much more time efficient ways of sorting tasks as compared to other methods like sorting lists on their own. This also makes retrieval of tasks much easier as we just need to extract the root node in order to get the task with the highest priority.

The second data structure used are classes. We have implemented three classes. One is the class that has the attributes and methods needed for the heap. The second is the Task class which takes the input and is compiled into a list. The Task class only has attributes and no functionality of its own as it does not do anything to the input. The third class is the TaskScheduler class which has the main functionality of the algorithm in the methods. The reason these two classes are kept separate is that they offer easy maintenance of the code, changes to the TaskScheduler class do not affect the way the input is given and the Task class makes it easier to change the way a certain input is percieved by the algorithm which was very useful especially when we did two different implementations of the greedy algorithm approach as it allowed versatility of approaches with respect to the utility value.

The only two differences that this algorithm has to the dynamic approach is that:

1. It does not make use of tables to store profit values as we are only foused on the profit/profit per minute value at each decision node.

2. It does not have the method that checks for the best possible profit value in the MultiTasker class.

#### Assumptions:

The first assumption in this case is that the user knows the profit that is to be gained out of each activity and the cost of each activity which is the main basis of this algorithm as without that, it won't work. This assumption is very important as a messing up on this part would mean that the user might be stuck doing tasks that are either too costly or do not have that much profit value thus quantifying both these factors is a must for the algorithm to work.

The second assumption is that the user has an understanding that not all tasks will be being done on the day due to the presence of the upper bound and some tasks will be skipped. The user has to identify this and make sure that no task is missed that is very essential e.g. attending a class, and if this task is being skipped, they should be re-evaluating their profits and costs for the inputs.

```python
#code for greedy programming scheduler 

class Task:
    """
    - id: Task Id   
    - description: Short description of the task   
    - duration: Duration in minutes   
    - dependencies: An array of id's of tasks that the task depends on 
    - status: Current status of the task
    - multi_tasking: Tells whether the task can be multitasked or not
    - fixed: Bool for if the task has a fixed start time or can start at any time within its urgency
    - priority: The urgency of the task 
   
    """
    
    
    def __init__(self,task_id,description,duration,dependencies, profit, cost = 0, fixed=0, multi_tasking=False, status="N"):
        self.id = task_id
        self.description = description
        self.duration = duration
        self.dependencies = dependencies
        self.status = status
        self.priority = round(((profit - cost)/duration), 5)
        self.multi_tasking = multi_tasking
        self.fixed = fixed
    
    #The format of the display
    def __repr__(self):
        return f"{self.description} - id: {self.id}\n \tDuration:{self.duration}\n\tDepends on: {self.dependencies}\n\tMust be completed at most {self.priority} minutes after day start\n\tFixed task start: {self.fixed}\n\tStatus: {self.status}"
    
    #Definining the comparison variable which in this case is the priority
    def __lt__(self, other):
        return self.priority < other.priority

    
class MultiTaskScheduler_Greedy1:
    """
    A Daily Task Scheduler Using Priority Queues based on priority that allows for multi-tasking
    Attributes
    ----------
    tasks : arr
        A Python list of all the tasks we schedule
    priority_queue: arr
        The priority queue of tasks ready to execute
    fixed_tasks: arr
        The priority queue of tasks that start at a fixed time
    """ 
    
    NOT_STARTED ='N'
    IN_PRIORITY_QUEUE = 'I'
    COMPLETED = 'C'
    
    def __init__(self, tasks):
        self.tasks = tasks
        self.priority_queue = MaxHeapq()
        self.fixed_tasks = MaxHeapq()
                
    def print_self(self):
        print('Input List of Tasks')
        for t in self.tasks:
            print(t)            
            
    def remove_dependency(self, task_id):
        """
        Input: list of tasks and id of the task completed
        Output: lists of tasks with completed task removed
        """
        for t in self.tasks:
            if t.id != task_id and task_id in t.dependencies:
                t.dependencies.remove(task_id)           
            
    def get_tasks_ready(self):
        """ 
        Input: list of tasks
        Output: list of tasks that are executable right now(no dependencies)
        """
        for task in self.tasks:
            if task.status == self.NOT_STARTED and not task.dependencies and not task.fixed: 
                task.status = self.IN_PRIORITY_QUEUE 
                self.priority_queue.heappush(task)
        
    def check_unscheduled_tasks(self):
        """
        Input: list of tasks 
        Output: boolean (checks the status of all tasks and returns True if at least one task has status = 'N'
        """
        for task in self.tasks:
            if task.status == self.NOT_STARTED:
                return True
        return False   
    
    def get_fixed_tasks(self):
        """
        Get all tasks in the task list that have fixed start times
        """
        for i in self.tasks:
            if i.fixed:
                self.fixed_tasks.heappush(i)
    
    def format_time(self, time):
        return f"{time//60}h{time%60:02d}"
    
    #method that checks for and executes multi-tasking
    def multi_tasker(self, task, current_time, profit):
        if (task.multi_tasking):
            temporary = []
            while(self.priority_queue.heap_size > 0 and (not self.priority_queue.maxk().multi_tasking or (self.fixed_tasks.heap_size > 0 and (self.priority_queue.maxk().duration > self.fixed_tasks.maxk().duration)))):
                temporary.append(self.priority_queue.heappop())

            # if there are tasks in the priority queue after the while loop is done, we can multi-task using them
            if self.priority_queue.heap:
                multitask = self.priority_queue.heappop()
                if ((current_time + task.duration)//60 < 24) and ((current_time + multitask.duration)//60 < 24):
                    print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing tasks {task.id} and {multitask.id} that take {task.duration} and {multitask.duration} mins respectively")
                
                #Increment the current time with the duration of the larger task(duration wise) as that task will still
                #be in process when the first task is completed.
                if task.duration < multitask.duration:
                    current_time += task.duration
                    if ((current_time + task.duration)//60 < 24) and ((current_time + multitask.duration)//60 < 24):
                        print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")
                        current_time += multitask.duration - task.duration
                        print(f"‚úÖ Completed Task {multitask.id} - '{multitask.description}' at time {self.format_time(current_time)}\n")
                        profit += (task.priority*task.duration) + (multitask.priority*task.duration)
                        
        
                else:
                    current_time += multitask.duration
                    if ((current_time + task.duration)//60 < 24) and ((current_time + multitask.duration)//60 < 24):
                        print(f"‚úÖ Completed Task {multitask.id} - '{multitask.description}' at time {self.format_time(current_time)}\n")
                        current_time += task.duration - multitask.duration
                        print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")
                        profit += (task.priority*task.duration) + (multitask.priority*task.duration)
                              
                
                self.remove_dependency(task.id)
                self.remove_dependency(multitask.id)
                task.status = self.COMPLETED
                multitask.status = self.COMPLETED

            #if the priority queue is empty after the while loop:
            #none of the tasks can be multitasked 
            #within the time between the current time and the start of the next fixed task
            #so we just perform the initial task without multi-tasking
            else:
                if ((current_time + task.duration)//60 < 24):
                    print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                    current_time += task.duration            
                    print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")  
                    profit += (task.priority*task.duration)
                    
                self.remove_dependency(task.id)
                task.status = self.COMPLETED
                

            #once we've carried out the task:
            #we move all the other tasks to the priority queue from the temporary list we had added them to
            for i in temporary:
                self.priority_queue.heappush(i)
            
        #if the task isn't multi-taskable we treat it as a regular task without running it through further comparisons
        else: 
            if (current_time + task.duration)//60 < 24:
                print(f"‚è∞Simple Scheduler at time {self.format_time(current_time)} started executing task {task.id} that takes {task.duration} mins")
                current_time += task.duration            
                print(f"‚úÖ Completed Task {task.id} - '{task.description}' at time {self.format_time(current_time)}\n")
                profit += (task.priority*task.duration)
                

            self.remove_dependency(task.id)
            task.status = self.COMPLETED
            
        #return the current time
        return current_time, profit
    
    def run_task_scheduler(self, starting_time = 480):
        current_time = starting_time
        profit = 0
        self.get_fixed_tasks()
        while self.check_unscheduled_tasks() or self.priority_queue.heap:
            self.get_tasks_ready()
            if current_time//60 > 24:
                print("End of the day. No more tasks possible today")
                break
            if self.priority_queue.heap_size > 0:      
                if self.fixed_tasks.heap_size > 0 and self.priority_queue.maxk().duration > self.fixed_tasks.maxk().duration:  
                    temporary = []
                    while(self.priority_queue.heap_size > 0 and self.priority_queue.maxk().duration > self.fixed_tasks.maxk().duration):
                        temporary.append(self.priority_queue.heappop())   
                    if self.priority_queue.heap:
                        task = self.priority_queue.heappop()  
                        # calling the multi_task method and update the current time to the 
                        #new time once the multi-task method is done.
                        current_time, profit = self.multi_tasker(task, current_time, profit)
                        print(profit)

                    else:
                        task = self.fixed_tasks.heappop()
                        current_time, profit = self.multi_tasker(task, current_time, profit)
                       
                    for i in temporary:
                        self.priority_queue.heappush(i)
                        
                #if the current task can be done before the next fixed task but can not be multi-tasked with another
                #task, we just perform it like a normal task.
                else:
                    task = self.priority_queue.heappop()   
                    current_time, profit = self.multi_tasker(task, current_time, profit)
                    
            else:
                while self.fixed_tasks.heap_size > 0:
                    current_time = current_time - self.fixed_tasks.maxk().duration
                    task = self.fixed_tasks.heappop()
                    current_time, profit = self.multi_tasker(task, current_time, profit)
                     
        total_time = current_time - starting_time             
        print(f"üèÅ Completed all planned tasks in {total_time//60}h{total_time%60:02d}min")
        print(f"Total profit = {profit}")
```

#### Time Complexity

The problem with these task schedulers as we can see is that the input that we give them is very hard to generate by hand even and doing it by random generation and trying to come up with a way to make an asymptotic notation would have been in vain. Thus, in order to discuss the time complexity of the dynamic multi tasker, we rely on time recurrence relations with respect to the logical reasoning of how the algorithm works.

Time recurrence calculation

Costs of various parts of the algorithm

Multi_tasker() = O(n) x O(n - 1) as the function is run on all the tasks and then run again on the tasks minus the task that was performed. This simplifies down to O(n^2) - O(n).

Run_taskscheduler() = O(n) x O(n - 1) as it runs the task and then runs the method again on the tasks that are left which also simplifies down to O(n^2) - O(n).

The methods within the class that are called such as get_tasks_ready(), and get_unscheduled_tasks() take O(n) time which repeated for a constant number of times will lead to C*O(n)

The runtime of the heappush() method is O(n log n) as it iterates the number of elements in the list it has to push, n and the time of insertion is log n.

The extraction of the root nodes for the sake of tasks take O(1) time and multiplied by a constant means the total time will be K*O(1)

Thus, the time recurrence equation becomes:

T(n) = O(n^2) - O(n) + O(n^2) - O(n) + CO(n) + O(n log n) + KO(1)

The 2O(n) + CO(n) will become C*O(n) as C is just a constant

T(n) = 2 O(n^2) + O(n log n) + CO(n) + KO(1)

As we reach asymptotic notation in theory, only the O(n^2) will be important and the others can be disregarded as their growth rate is too small for significant impact. Moreover, the constant, 2 which is the multiple of O(n^2) can also be disregarded as it is just a constant multiple.

Thus, the final time recurrence equation becomes:

T(n) = O(n^2)

This means that each task added to the scheduler will result in a n^2 factor increase in run time because when we add a task, the program will have an n^2 number of increase in the comparisons/iterations that it has to do over the task.
